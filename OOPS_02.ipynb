## 1. Basics of Object-Oriented Programming


- Problem 1: Create a class named Ingredient . The constructor should take
ingredient_id , name , and unit_cost_inr as arguments. The class should also have
a method named get_ingredient_details that returns a formatted string with the
ingredient's ID, name, and unit cost.


```python
class Ingredient():
    def __init__(self , ingredient_id , name , unit_cost_inr):
        self.ingredient_id = ingredient_id
        self.name = name
        self.unit_cost_inr = unit_cost_inr
    
    def get_ingredient_details(self):
        return f"Ingredient ID: {self.ingredient_id}, Name: {self.name}, Unit Cost: {self.unit_cost_inr} INR"

```

- Problem 2: Create a new class named Recipe . This class should have a constructor that
accepts recipe_name and ingredients_needed (a dictionary where keys are
Ingredient objects and values are the quantity needed). Add a method named
calculate_total_cost that computes the total cost of the recipe based on the
quantity and unit cost of each ingredient. Create an Ingredient object for "milk" and
"sugar" and then a Recipe object for "Vanilla Dessert Base".


```python



class Recipe():
    def __init__(self , recipe_name , ingredients_needed):
        self.recipe_name = recipe_name
        self.ingredients_needed = ingredients_needed
    
    def calculate_total_cost(self):
        total_cost = 0
        for ingredient, quantity in self.ingredients_needed.items():
            total_cost += ingredient.unit_cost_inr * quantity
        return total_cost
    
milk = Ingredient("ING-001" , "Milk" , 50)
sugar = Ingredient("ING-002" , "Sugar" , 30)
print(milk.get_ingredient_details())
print(sugar.get_ingredient_details())

Vanilla_dessert_base = Recipe("Tea" , {milk: 2 , sugar: 1})
print(Vanilla_dessert_base.calculate_total_cost())
```

- Problem 3: Add a method to the Recipe class called is_profitable . This method
should take a selling_price_inr as an argument and return True if the selling price
is more than twice the total cost, and False otherwise. Use this method to check if the
"Vanilla Dessert Base" recipe is profitable if its selling price is 500 INR.

```python
class Recipe():
    def __init__(self , recipe_name , ingredients_needed):
        self.recipe_name = recipe_name
        self.ingredients_needed = ingredients_needed
    
    def calculate_total_cost(self):
        total_cost = 0
        for ingredient, quantity in self.ingredients_needed.items():
            total_cost += ingredient.unit_cost_inr * quantity
        return total_cost

    def is_profitable(self,selling_price_inr):
        total_cost = self.calculate_total_cost()
        if selling_price_inr > total_cost:
            return True
        else:
            return False
        
Vanilla_dessert_base = Recipe("Tea" , {milk: 2 , sugar: 1})
print(Vanilla_dessert_base.is_profitable(500))
```

## 2. Object-Oriented Programming using Inheritance


- Problem 1: Create a base class named Worker with a constructor that takes
employee_id , name , and department as arguments. Add a method named
show_basic_info that prints the worker's ID and name.

```python
class Worker():
    def __init__(self,employee_id , name , department):
        self.employee_id = employee_id
        self.name = name
        self.department = department
    def show_basic_info(self):
        print(f"Employee ID: {self.employee_id}, Name: {self.name} , Department: {self.department}")
```

- Problem 2: Create a subclass named ProductionWorker that inherits from Worker .
The constructor should also accept shift and hourly_rate_inr . It should use the
super() function to call the parent class constructor and should have its own method to
calculate daily wages based on an 8-hour shift.


```python
class ProductionWorker(Worker):
    def __init__(self , shift , hourly_rate_inr , employee_id , name , department):
        super().__init__(employee_id , name , department)
        self.shift = shift
        self.hourly_rate_inr = hourly_rate_inr
    
    def calculate_daily_wages(self):
        return self.hourly_rate_inr * 8
```

```python
class QualityController(Worker):
    def __init__(self ,certification_id , employee_id , name , department):
        super().__init__(employee_id , name , department)
        self.certification_id = certification_id
    
    def inspect(self,status):
        if status == 'Pass':
            print("Product batch passed the quality check.")
        else:
            print("Product batch failed the quality check.")
```

- Problem 5: In the Payment class, add a new method called apply_discount . This
method should take one argument, discount_percentage . If no argument is provided,
it should default to a 5% discount. Show an example of calling this method with and
without the argument to demonstrate method overloading (using default arguments).

```python
class Payment:
    def __init__(self, amount):
        self.amount = amount
        self.final_amount = amount

    # The method uses a default argument for method overloading demonstration
    def apply_discount(self, discount_percentage=5):
        # Convert percentage to a decimal (e.g., 5 becomes 0.05)
        discount_factor = discount_percentage / 100
        
        # Calculate the final amount after discount
        self.final_amount = self.amount * (1 - discount_factor)
        
        print(f"Applying a {discount_percentage}% discount.")
        print(f"Original: ${self.amount:.2f}, Final: ${self.final_amount:.2f}")

# --- Demonstration of Method Overloading via Default Arguments ---
print("--- Example 1: Calling WITHOUT Argument (Default 5%) ---")
# Creates payment object with an initial amount of 100.00
payment_A = Payment(100.00) 
# Call the method without an argument; it uses the default value of 5
payment_A.apply_discount() 
print(f"Check: payment_A final amount is ${payment_A.final_amount:.2f}\n")

print("--- Example 2: Calling WITH Argument (Custom 20%) ---")
# Creates a second payment object with an initial amount of 50.00
payment_B = Payment(50.00) 
# Call the method with a custom argument (20)
payment_B.apply_discount(20) 
print(f"Check: payment_B final amount is ${payment_B.final_amount:.2f}")
```

- Problem 1: Create an abstract base class named BaseProduct using the abc module.
This class should have an abstract method named get_price and another abstract

```python
from abc import ABC , abstractmethod

class BaseProduct(ABC):
    @abstractmethod
    def get_price(self):
        pass

    @abstractmethod
    def calculate_profit(self):
        pass
```

## 5. Dunder methods Object-Oriented Programming


- Problem 1: Create a class named StockItem with a constructor that takes item_name ,
quantity , and reorder_level . Implement the __repr__ dunder method so that
when you inspect the object, it returns a string that can be used to recreate it, like
"StockItem('Mango Puree', 100, 50)".

```python
class StockItem():
    def __init__(self,item_name , quantity , reorder_level):
        self.item_name = item_name
        self.quantity = quantity
        self.reorder_level = reorder_level
    
    def __repr__(self):
        return f"StockItem('{self.item_name}', {self.quantity}, {self.reorder_level})"
```

- Problem 2: In the StockItem class, implement the __lt__ dunder method to allow
you to compare two StockItem objects. The comparison should be based on their
quantity. Create two StockItem objects and use the < operator to check if one has a
lower quantity than the other. Print the result.


```python
class StockItem():
    def __init__(self,item_name , quantity , reorder_level):
        self.item_name = item_name
        self.quantity = quantity
        self.reorder_level = reorder_level
    
    def __repr__(self):
        return f"StockItem('{self.item_name}', {self.quantity}, {self.reorder_level})"
    
    def __lt__(self , other):
        return self.quantity < other.quantity
```

- Problem 3: Implement the __bool__ dunder method in the StockItem class. This
method should return True if the item's quantity is above its reorder level and False
otherwise. Use the bool() function on a StockItem object to test its stock status.

```python
class StockItem():
    def __init__(self,item_name , quantity , reorder_level):
        self.item_name = item_name
        self.quantity = quantity
        self.reorder_level = reorder_level
    
    def __repr__(self):
        return f"StockItem('{self.item_name}', {self.quantity}, {self.reorder_level})"
    
    def __lt__(self , other):
        return self.quantity < other.quantity
    
    def __bool__(self ):
       return self.quantity > self.reorder_level
          
A = StockItem("Mango Puree", 100, 50)
B = StockItem("Apple Puree", 30, 50)

print(bool(A))  
print(bool(B)) 
```
